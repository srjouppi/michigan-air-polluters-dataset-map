```python
import pandas as pd
```

```python
df1 = pd.read_csv("csv/EGLE-AQD-documents.csv")
df2 = pd.read_csv("csv/EGLE-AQD-source-list-final.csv")
```

```python
df1['datetime'] = pd.to_datetime(df1.date.astype(str), format = "%Y%m%d", errors = 'coerce')
```

```python
df1['date'] = df1['datetime']
df1 = df1.drop(['datetime'],axis=1)
df1
```

```python
df2['zip_code'] = df2['zip_code'].astype(str).str[:5]
df2
```

```python
# Let's make sure we get everything just for good measure
all_docs = df2.merge(df1, how='outer', left_on='id', right_on='source_id',)
all_docs
```

```python
all_docs[all_docs.id.isnull()].head()
```

```python
all_docs[all_docs.date.isnull()].head()
```

```python
# For practicality's sake let's just keep the records that have documents and IDs
most_docs = df1.merge(df2, right_on='id', left_on='source_id')
most_docs = most_docs.drop(['id'], axis=1)
```

```python
most_docs
```

```python
# Let's save this
most_docs.to_csv("csv/EGLE-AQD-documents-source-info.csv", index=False)
```

```python
# Filtering for Violation Notices
vn = most_docs[most_docs.doc_type == 'VN']
```

```python
vn = vn.drop(['id'],axis=1)
```

```python
vn.to_csv("csv/EGLE-AQD-violation-notices.csv",index=False)
```
