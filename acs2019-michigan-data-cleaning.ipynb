```python
import pandas as pd
# Reading in my census data -- race by zip code
# Source: https://censusreporter.org/data/table/?table=B02001&geo_ids=04000US26,860|04000US26&primary_geo_id=04000US26#
df = pd.read_csv("csv/acs2019-race-zip-michigan-raw.csv")
```

```python
df
```

```python
# Getting a percentage of the total for each column except population, name and geojson info
cols = ['B02001001, Error','B02001002', 'B02001002, Error', 'B02001003', 'B02001003, Error','B02001004','B02001004, Error','B02001005', 'B02001005, Error','B02001006','B02001006, Error','B02001007','B02001007, Error','B02001008','B02001008, Error','B02001009','B02001009, Error','B02001010','B02001010, Error']
df[cols] = df[cols].div(df['B02001001'], axis=0).multiply(100).round(1)
```

```python
df
```

```python
# Renaming the columns
df = df.rename(columns={ "B02001001": "Total", "B02001002": "white_pct", "B02001003": "black_pct", "B02001004": "am_indian_ak_native_pct", "B02001005": "asian_pct", "B02001006": "hi_pc_islander_pct", "B02001007": "other_alone_pct", "B02001008": "two_more_pct", "B02001009": "two_other_pct", "B02001010": "two_three_pct"}, errors="raise")
```

```python
# Renaming the error columns
df = df.rename(columns={ "B02001001, Error": "Total, Error", "B02001002, Error": "white_err", "B02001003, Error": "black_err", "B02001004, Error": "am_indian_ak_native_err", "B02001005, Error": "asian_err", "B02001006, Error": "hi_pc_islander_err", "B02001007, Error": "other_alone_err", "B02001008, Error": "two_more_err", "B02001009, Error": "two_other_err", "B02001010, Error": "two_three_err"}, errors="raise")
df
```

```python
# Adding up the other columns
df['other_pct'] = df['other_alone_pct'] + df['two_more_pct']
df
```

```python
# Does everything add up?
# Seems like it, I think because of error, it might not add up always perfectly to 100
# would check with Census data expert on this before publishing
df['total_check'] = df['white_pct'] + df['black_pct'] + df['am_indian_ak_native_pct'] + df['asian_pct'] + df['hi_pc_islander_pct'] + df['other_pct']
```

```python
df
```

```python
# Making a copy with the columns that I want and exporting it to a csv
df[['name','white_pct', 'black_pct','am_indian_ak_native_pct','asian_pct','hi_pc_islander_pct','other_pct']].copy().to_csv("csv/acs2019-race-zip-michigan-clean.csv",index=False)
```

```python
# Reading in my census data --- poverty
# Source: https://censusreporter.org/data/table/?table=B17001&geo_ids=04000US26,860|04000US26&primary_geo_id=04000US26
df2 = pd.read_csv("csv/acs2019-poverty-zip-michigan-raw.csv")
```

```python
# Making a copy with the columns I want
df2 = df2[['name', 'B17001001', 'B17001001, Error','B17001002','B17001002, Error']].copy()
df2
```

```python
# Renaming the columns
df2 = df2.rename(columns={"B17001001":"Total","B17001001, Error": "Total, Error", "B17001002": "Income Poverty Level","B17001002, Error": "Income Poverty Level, Error"})
```

```python
df2
```

```python
# Getting percentage of the total
# Exporting to csv
cols = ['Total, Error', 'Income Poverty Level', 'Income Poverty Level, Error']
df2[cols] = df2[cols].div(df2['Total'], axis=0).multiply(100).round(1)
df2.to_csv("csv/acs2019-poverty-zip-michigan-clean.csv")
```
