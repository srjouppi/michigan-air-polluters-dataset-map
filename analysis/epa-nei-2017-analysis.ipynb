```python
import pandas as pd
```

```python
# The document is large. You can download "2017neiJan_facility.zip" here: https://gaftp.epa.gov/air/nei/2017/data_summaries/2017v1/
# If your computer has a hard time unzipping the large file, google 'unzip large file with python' or command line
df = pd.read_csv("INSERT PATH TO EPA NEI CSV")
```

```python
pd.set_option("max_columns", None)
df.head()
```

```python
# Let's look at the most common pollutants
df['pollutant code'].value_counts().head(30)
```

```python
# Filtering data to Michigan
df = df[df.state == 'MI']
```

```python
# Looking at number of facilities emitting Sulfur Dioxide by County
df[df['pollutant code'] == "SO2"].county.value_counts()
```

```python
# Checking the time period
# Seems to all be from reporting period A.
df['reporting period'].value_counts()
```

```python
# Getting a data frame of Hazardous Air Polutants by facility ID so I can merge this with my dataframe
hap_count = df[df['pollutant type(s)'] == 'HAP'].groupby('agency facility id')['pollutant type(s)'].value_counts().to_frame()
hap_count
```

```python
hap_count.to_csv('csv/hap-by-source-id.csv')
```

```python
# Looking at total of Sulfur Dioxide emissions by county
df[df['pollutant code'] == "SO2"].groupby(by='county')['total emissions'].sum().sort_values(ascending = False)
```

```python
# Looking at total of noxious gas emissions by county
df[df['pollutant code'] == "NOX"].groupby(by='county')['total emissions'].sum().sort_values(ascending = False)
```

```python
# Looking at the total of Particulate Matter 2.5 emitted by county
# !!! WAYNE COUNTY !!!
df[df['pollutant code'] == "PM25-PRI"].groupby(by='county')['total emissions'].sum().sort_values(ascending = False)
```

```python
# Looking at Arsenic emissions by county
df[df['pollutant desc'] == "Arsenic"].groupby(by='county')['total emissions'].sum().sort_values(ascending = False)
```

```python
# Which sources are the worst emitters of particulate matter?
df[df['pollutant code'] == "PM25-PRI"].sort_values(by = 'total emissions', ascending = False)
```

```python
# Which facilities emit the most Arsenic?
df[df['pollutant desc'] == "Arsenic"].sort_values(by = 'total emissions', ascending = False)
```

```python
# Which facilities emit the most Hazardous Air Pollutants?
# Holy COW Rieth Riley Construction shows up 3 times so far...
df[df['pollutant type(s)'] == 'HAP'].groupby(by='site name')['pollutant type(s)'].value_counts().sort_values(ascending = False)
```
