```python
import pandas as pd
```

```python
# Importing camelot and reading the PDF
import camelot
import re
tables = camelot.read_pdf("docs/Sources_By_ZIP.pdf",
                          flavor='stream',
                          pages='1-end')
```

```python
# Making a super dataframe by concatting all of the dataframes
dfs = [table.df for table in tables]
df = pd.concat(dfs, ignore_index=True)
df.to_csv("csv/MI-source-list-converted.csv", index=False)
```
